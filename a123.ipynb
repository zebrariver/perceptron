```python
#导入函数库
import numpy as np
import matplotlib.pyplot as plt
#定义样本标签数
num=50
#随机数种子
np.random.seed(0)
#随机生成数据样本
x=np.r_[np.random.randn(num,2)+[1,1],np.random.randn(num,2)-[1,1]]
#标签
y=[-1]*num+[1]*num
#画图
plt.figure()
for i in range(len(x)):
    if y[i]==1:
        plt.plot(x[i][0],x[i][1],'ro')
    else:
        plt.plot(x[i][0],x[i][1],'bo')
```

```python
#定义初始权值、偏置、学习率
w=np.array([-1,1])
b=0
delta=1e-3
#重绘散点图
for i in range(len(x)):
    if y[i]==1:
        plt.plot(x[i][0],x[i][1],'ro')
    else:
        plt.plot(x[i][0],x[i][1],'bo')
#开始训练        
for i in range(101):
    grad_x=np.array([0,0])
    grad_b=0        #初始化梯度值
    for j in range(len(x)):
        if y[j]!=np.sign(np.dot(w,x[j])+b):   #寻找错分样本
            grad_x=grad_x+y[j]*x[j]
            grad_b=grad_b+y[j]
    w=w+delta*grad_x
    b=b+delta*grad_b
if i%10==0:
        xx=[-4,4]
        yy=[0,0]
        for i in range(len(xx)):
            yy[i]=-(w[0]*xx[i]+b)/w[1]
        pp=plt.plot(xx,yy)

```
